## Webscrapping

```python
def url_bs4(url):
    get_html = requests.get(url)
    print(get_html.status_code)
    print(get_html.encoding)
    html = get_html.content
    soup = BeautifulSoup(html)
    return soup
```

```python
url = 'https://pt.wikipedia.org/wiki/Lista_da_programa%C3%A7%C3%A3o_do_Lollapalooza_por_ano'
soup_wiki = url_bs4(url)
```

```python
wiki_a_tag = soup_wiki.select(".wikitable a")
```

```python
wiki_a_tag[141]
```

```python
# converter cada objeto Tag em uma string
lista_strings = [str(tag) for tag in wiki_a_tag]

pattern = r'href="([^"]*)" title="([^"]*)"'

# criar listas para armazenar valores extrados
href_values = []
title_values = []

# iterar sobre cada string na lista e extrair valores com regex
for string in lista_strings:
    match = re.search(pattern, string)
    if match:
        href_values.append(match.group(1))
        title_values.append(match.group(2))

# criar dicionrio com as listas de valores
data = {"href": href_values, "title": title_values}

# criar dataframe a partir do dicionrio
df_wiki = pd.DataFrame(data)

# imprimir o dataframe
df_wiki
```

```python
def preenchimento_link(row):
    if "pgina no existe" in row["title"]:
        return np.nan
    else:
        return 'www.wikipedia.org'+ row["href"]
        
updated_links = df_wiki.apply(preenchimento_link, axis = 1)
```

```python
df_wiki['wiki_pages'] = updated_links
```

```python
df_wiki = df_wiki.drop("href", axis=1)
```

```python
def artista_uri(nome_artista):
    """Procura o nome do artista no spotify e retorna o nmero identificador(URI) desse artista"""
    try:
        result_artista = spotify.search(q=nome_artista,type='artist')
        df_artistas = pd.json_normalize(result_artista['artists']['items'])
        mask_artistas = df_artistas['name'].str.title() == nome_artista.title()
        uri = df_artistas[mask_artistas].iloc[0,7]
        return uri
    except:
        return np.nan
```

```python
def artist_info_spotify(array):
    artist_info = []
    for n in array:
        resultado_artista = spotify.artist(n) 
        name = resultado_artista["name"]
        followers = resultado_artista["followers"]["total"]
        genres = resultado_artista["genres"]
        href = resultado_artista["href"]
        id_spotify = resultado_artista["id"]
        popularity = resultado_artista["popularity"]
        artist_info_single = pd.DataFrame([name,followers, genres, href, id_spotify, popularity],index=['name','followers', 'genres', 'href', 'id', 'popularity']).T
        artist_info.append(artist_info_single)
        
    df_artist_info = pd.concat(artist_info, ignore_index = True)
    return df_artist_info
```

```python
def top_musicas(uri_artista):
    """Recebe a uri do artista e retorna um dataframe com as top 10 musicas e as caractericas de audio delas """
    try:
        top_tracks = spotify.artist_top_tracks(uri_artista)

        df_music = pd.json_normalize(top_tracks['tracks'])
        musicas = df_music['uri'].tolist()
        df_music_caracteristicas = pd.json_normalize(spotify.audio_features(musicas))
        df_music_final = pd.merge(df_music,df_music_caracteristicas,on='uri')
    
        return df_music_final
    except:
        pass
```

```python
musicas_lista = []
for uri in array_id:
    music_info = top_musicas(uri)
    musicas_lista.append(music_info)
        
df_top_tracks= pd.concat(musicas_lista, ignore_index = True)
```

```python
def related_artist_info_spotify(array):
    related_artist_info = []
    for n in array:
        resultado_artista = spotify.artist_related_artists(n)
        for i in range(len(resultado_artista['artists'])):
            id_original_artist = n
            name = resultado_artista['artists'][i]["name"]
            followers = resultado_artista['artists'][i]["followers"]["total"]
            genres = resultado_artista['artists'][i]["genres"]
            uri = resultado_artista['artists'][i]["uri"]
            id_spotify = resultado_artista['artists'][i]["id"]
            popularity = resultado_artista['artists'][i]["popularity"]
            related_artist_info_single = pd.DataFrame([id_original_artist, name,followers, genres, uri, id_spotify, popularity],index=['id_original_artist','name','followers', 'genres', 'uri', 'id', 'popularity']).T
            related_artist_info.append(related_artist_info_single)
        
    df_related_artist_info = pd.concat(related_artist_info, ignore_index = True)
    return df_related_artist_info
```
